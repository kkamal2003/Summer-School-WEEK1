```python
import matplotlib.pyplot as plt
Filesize = [200,400,600,800,1000]
c = [12,20,34,45,55]
c1 = [15,25,36,50,60]
j = [18,30,40,55,70]
R = [20,35,45,60,80]
p = [25,40,53,75,100]
plt.plot(Filesize,c,label='C');
plt.plot(Filesize,c1,label='C++');
plt.plot(Filesize,j,label='Java');
plt.plot(Filesize,R,label='R');
plt.plot(Filesize,p,label='Python');
plt.legend();
plt.xlabel('Filesize');
plt.ylabel('Time Taken (sec)');
plt.title('Performance Comparsion')
plt.show()
```
